```python
import pandas as pd
from spellchecker import SpellChecker
import nltk
from nltk import word_tokenize
import re
```

# **Pyspell checker**

*   It uses a **Levenshtein Distance** algorithm to find permutations within an edit distance of 2 from the original word. It then compares all permutations (insertions, deletions, replacements, and transpositions) to known words in a word frequency list. Those words that are found more often in the frequency list are more likely the correct results.





# **DATASET**


> In this task, Book.txt dataset was taken for auto correct.This data set contains 19052 sentences.

> Here we use first **100 sentences and 200 words** for autocorrection.first we download this dataset through online and create some spell mistake on this dataset.

> Now we check the misspelled words into correct words using **spell checker module**


# **1) Importing  Dataset**

```python
df=pd.read_fwf('book.txt')
df.columns=['sentences']
df.head(10)
```

# **2) Cleaning(Preprocessing) the data**

```python
def clean(data,column):
  data=data.drop_duplicates()
  data['new_sen']=[re.sub('[0-9]','',i) for i in data[column]]
  data.dropna()
  data['new_sen']=data[column].str.replace("[^a-zA-Z]"," ")
  return data
```

```python
df=clean(df,'sentences')
df.head(20)
```

```python
nltk.download('all')
```

```python
from nltk.corpus import stopwords
stop_words=stopwords.words('english')
```

```python
def remove_stop(sentence,stop_words):
  rem_stop=[]
  token=nltk.word_tokenize(sentence)
  for j in token:
    j=j.lower()
    if len(j)>3:
      if j not in stop_words:
        rem_stop.append(j)
  new_rev=' '.join(rem_stop)
  return new_rev
```

```python
df['new_sen']=[remove_stop(sen,stop_words) for sen in df['new_sen']]
df
```

#**3)Auto correction of  100 sentences**

```python
def autocorrection(sentence):
  tokens=nltk.word_tokenize(sentence)
  spell=SpellChecker()
  new_sen=[]
  for i in tokens:
    if spell.correction(i.lower())==None:
      new_sen.append(i)
    else:
      new_sen.append(spell.correction(i.lower()))
  return " ".join(new_sen)
```

```python
df_new=df.iloc[:100,0:]
df_new['correct sentences']=[autocorrection(sen) for sen in df_new['new_sen']]
df_new
```

# **4)Auto correction on 200 words**

```python
words=[]
for i in df['new_sen']:
  for j in i.split():
    words.append(j.lower())
len(words)
```

```python
df=pd.DataFrame(words)
df.columns=['words']
```

```python
df=df.drop_duplicates()
df.shape
```

```python
def autocorrect(word):
  spell=SpellChecker()
  if spell.correction(word.lower())==None:
    return word
  else:
    return spell.correction(word.lower())
```

```python
new_words=df['words'][:200]

len(new_words)
```

```python
cor_words=[autocorrect(i) for i in df['words'][:200]]
cor_words
```

```python
df_200words=pd.DataFrame(list(zip(new_words,cor_words)),columns=['first 200 words','First 200 correct words'])
df_200words.head(50)
```

# **5) Suggestions of word**

```python
def suggestions(word):
  spell=SpellChecker()
  if spell.candidates(word.lower())==None:
    return word
  else:
    return spell.candidates(word.lower())
```

```python
suggestions=[suggestions(i) for i in new_words]
suggestions
```

```python
df_200words=pd.DataFrame(list(zip(new_words,suggestions,cor_words)),columns=['first 200 words','suggestions of words','First 200 correct words'])
df_200words
```

# **1) Here the first column is given word for auto correct test**

# **2) Second column gives suggestion of misspelled words for given word.**

# **3) Final column gives a most suitable word for all among the suggestions comparing the probabality of suggested words.**
